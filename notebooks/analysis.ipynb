```python
from icxcca import *
import matplotlib.pyplot as plt

data_filepath = "../data/npz/2022_04_14_0002_trunc.npz"

dm = DataManager(data_filepath)
# _ = dm.plot()
```

```python
v_traces = dm.get_voltage_data()
# plot_spike_info(v_traces['Voltage 1'])

# plot one trace and its second derivative
trace = v_traces['Voltage 1']
plt.plot(trace.t, trace.d)
plt.show()

second_derivative = trace.derivative().derivative()
plt.plot(second_derivative.t, second_derivative.d)
plt.show()

```
